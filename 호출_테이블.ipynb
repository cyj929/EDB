```python
import geopandas as gpd
from shapely.geometry import Point
import random
import pandas as pd
import matplotlib.pyplot as plt
from pyproj import Proj, transform

# Load your shapefile
shapefile_path = '/content/LSMD_CONT_LDREG_11500.shp'
gdf = gpd.read_file(shapefile_path)

# Define the number of random points to generate
num_points = 200

# Generate random points within the polygons and store their coordinates
random_points = []
random_x_coordinates = []
random_y_coordinates = []

# Get the EPSG code of the shapefile CRS
utm_crs = gdf.crs.to_epsg()

for _ in range(num_points):
    while True:
        # Generate random coordinates within the bounds of the entire shapefile
        minx, miny, maxx, maxy = gdf.total_bounds
        x = random.uniform(minx, maxx)
        y = random.uniform(miny, maxy)
        point = Point(x, y)

        # Check if the point is within any of the polygons in the shapefile
        if gdf.geometry.contains(point).any():
            random_points.append(point)
            random_x_coordinates.append(x)
            random_y_coordinates.append(y)
            break

# Convert the list of random points to a GeoDataFrame
random_points_gdf = gpd.GeoDataFrame(geometry=random_points)

# Create a DataFrame with x and y coordinates
data = {'X': random_x_coordinates, 'Y': random_y_coordinates}
df = pd.DataFrame(data)

# Save DataFrame to Excel
excel_path = 'random_points_coordinates.xlsx'
df.to_excel(excel_path, index=False)

# Convert the x and y coordinates to latitude and longitude
in_proj = Proj(init=f'epsg:{utm_crs}')
out_proj = Proj(init='epsg:4326')  # WGS 84 (latitude, longitude)

df['Longitude'], df['Latitude'] = transform(in_proj, out_proj, df['X'].values, df['Y'].values)

# Print the latitude and longitude coordinates of the random points
for i, (lat, lon) in enumerate(zip(df['Latitude'], df['Longitude']), 1):
    print(f"Random Point {i}: Latitude={lat}, Longitude={lon}")

# Plot the shapefile and random points
ax = gdf.plot()
random_points_gdf.plot(ax=ax, color='red', markersize=10)
plt.show()

print(f"\nExcel file saved at: {excel_path}")
```

```python
import pandas as pd
from pyproj import Proj, transform

#   
excel_file_path = '/content/random_points_coordinates (2).xlsx'

#     
source_proj = Proj(init='epsg:5186')  #  
target_proj = Proj(init='epsg:4019')  # GRS80 

#   
df = pd.read_excel(excel_file_path)

#    
def transform_coordinates(row):
    x = row['X']  # 'X'   X   
    y = row['Y']  # 'Y'   Y   
    grs80_x, grs80_y = transform(source_proj, target_proj, x, y)
    return pd.Series({'GRS80_X': grs80_x, 'GRS80_Y': grs80_y})

#   
df[['GRS80_X', 'GRS80_Y']] = df.apply(transform_coordinates, axis=1)

#    
output_excel_path = 'output_coordinates.xlsx'
df.to_excel(output_excel_path, index=False)

print(f"  .  {output_excel_path} .")
```

```python
import pandas as pd
import random

# Load the Excel file into a pandas DataFrame
file_path = '/content/ .xlsx'  # Replace with your file path
df = pd.read_excel(file_path)

# Define the probabilities for each destination type
probabilities = {
    '': 0.3637,
    '': 0.1379,
    '': 0.0739,
    '': 0.0652,
    '': 0.3693
}

# Select addresses based on destination type
selected_addresses = []

# Calculate the total number of addresses to be sampled
total_addresses = 3000

for dest_type, prob in probabilities.items():
    dest_type_addresses = df[df[' '] == dest_type]

    # Calculate the number of addresses to sample based on probability
    num_addresses = min(int(prob * total_addresses), len(dest_type_addresses))

    # Randomly sample addresses for the current destination type without replacement
    sampled_addresses = dest_type_addresses.sample(n=num_addresses, replace=False)

    # Append the sampled addresses to the list
    selected_addresses.append(sampled_addresses)

# Concatenate the sampled addresses into a single DataFrame
result_df = pd.concat(selected_addresses)

# Calculate the remaining number of addresses to be sampled
remaining_addresses = total_addresses - len(result_df)

# Randomly sample the remaining addresses from all destination types
remaining_addresses_df = df.sample(n=remaining_addresses, replace= True)

# Concatenate the remaining addresses to the result DataFrame
result_df = pd.concat([result_df, remaining_addresses_df])

# Shuffle the DataFrame to randomize the order
result_df = result_df.sample(frac=1).reset_index(drop=True)

# Display the resulting DataFrame or save it to a new Excel file
print(result_df)
result_df.to_excel('output_addresses.xlsx', index=False)  # Uncomment this line to save the result to a new Excel file
```
